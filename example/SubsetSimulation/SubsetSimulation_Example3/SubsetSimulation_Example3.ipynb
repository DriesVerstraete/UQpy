This example runs subset simulation with the affine invariant ensemble sampler with stretch moves for MCMC to estimate the probability of failure of a system with linear performance function and 2-dimensional standard unit normal random variables, $U_i$ given by:

$ g(U) = \beta\sqrt{n} -\sum_{j=1}^2U_i$

$\beta=3$ corresponding to a true probability of failure $p_f=\Phi(-3)=0.0013$.

In this example, initial samples, i.e. level 0, are drawn a priori as standard normal Monte Carlo samples. Alternatively, the initial samples can be drawn internally in the SubsetSimulation class. This is illustrated in SubsetSimulation_Example1.ipynb and SubsetSimulation_Example1.ipynb.

Import the necessary libraries

```python
from UQpy.Reliability import SubsetSimulation
import matplotlib.pyplot as plt
import numpy as np
import time
```

Generate the initial samples as standard normal random variables.

```python
x_init = np.random.normal(size=(1000,2))
```

Here, we use 1000 samples per subset and call the performance function using a Python script (provided) called 'pfn.py' that evaluates $g(U)$ as defined above. The conditional probabilities for each level are 

```python
t = time.time()
x_ss = SubsetSimulation(dimension=2, samples_init=x_init, nsamples_ss=1000, pdf_proposal_scale=2, algorithm='Stretch', 
                        model_type='python', model_script='pfn.py', p_cond=0.1)
t_run = time.time()-t
```

Print the output attributes of x_ss.

```python
print(t_run)
print(x_ss.pf)
print(x_ss.g_level)
```

Plot the samples.

```python
for i in range(len(x_ss.g_level)):
    plt.plot(x_ss.samples[i][:,0],x_ss.samples[i][:,1],'x')
plt.show()
```
